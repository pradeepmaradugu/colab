```python
#What is pandas groupby? Demonstrate with your own example
import pandas as pd
df = pd.DataFrame({'Animal': ['tiger', 'tiger','fox', 'fox'],
                   'Max Speed': [380, 370, 120, 26]})

df.groupby(['Animal']).mean()
```

```python
#What are the ways to combine the dataframe in pandas. Give an example for each
#concat

dict1 = {
  'brand': ['ford','hyundai','tata','maruthi'],
  'model': ['mustang','i10','nexon','brezza'],
  'year': [1980,1995,2018,2012]
}
df1 = pd.DataFrame(dict1)
dict2 = {
   'brand': ['ford','hyundai','tata','maruthi'],
  'electric':[True, False,True,False],
    'colors': ["red", "white", "blue","black"]}

df2 = pd.DataFrame(dict2)

pd.concat([df1,df2])
```

```python
pd.merge(df1,df2,on=['brand','brand'])
```

```python
#Find out the mean, median and standard deviation of this numpy array -> np.array([1,5,3,100,4,48])
import numpy as np
a1 = np.array([1,5,3,100,4,48])

df = pd.DataFrame(a1)
df.agg([np.mean,np.median,np.std])
```

```python
#What is a map() function in Python? Give an example
#The map() function applies a given function to each item of an iterable (list, tuple etc.) and returns an iterator.

numb = [2,3,4,5,6]

def sq(n):
    return n*n

sq_numb = list(map(sq,numb))
sq_numb
```
