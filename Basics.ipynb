```python
def abs_func(v):
    if v >0:
        print(v)
    else:
        abs_value = abs(v)
        print(abs_value)
```

```python
abs_func(-8)
```

```python
def abs_simple(v):
    if v > 0:
        return v
    else:
        return -v
```

```python
abs_simple(-8)
```
